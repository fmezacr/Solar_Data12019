```python
import numpy as np #Importa Numpy
import peakutils #Importa Peakutils
from peakutils.plot import plot as pplot #Importa el modulo plot de Peakutils
from matplotlib import pyplot #Importa el modulo pyplot de Matplotlib
%matplotlib inline
```

```python
centers = (30.0, 70.0) #Crea una tupla con los valores centrales
x = np.linspace(0, 500, 150) #Crea una arreglo espaciado uniformemente espaciado que va de 0 a 500 con 150 muestras
y = (peakutils.gaussian(x, 5, centers[0], 3) +  
    peakutils.gaussian(x, 7, centers[1], 10) +
    np.random.rand(x.size)) 
#La linea anterior calcula 2 funciones Gaussianas; la primera en el intervalo de x con amplitud de 5, centrado en 30.0
#y un ancho de 3. La segunda se calcula en el intervalo de x con amplitud de 7, centrado en 70.0 y un ancho de 10. Se le
#suma un arreglo de numeros random del tamaño de x para agregar ruido a la figura
pyplot.figure(figsize=(10,6)) #Declara una figura con el tamaño establecido
pyplot.plot(x, y) #Plotea la figura usando los datos
pyplot.title("Datos con ruido") #Le pone titulo a la figura
pyplot.grid() #Activa la cuadricula en la figura
```

```python
indexes = peakutils.indexes(y, thres=0.3, min_dist=10) #El metodo peakutils.indexes() permite obtener los indices de los
#picos en los datos. Sus parametros son un arreglo con la amplitud de los datos, un umbral normalizado (threshold) para
#detectar picos con una amplitud mayor al umbral y una distancia minima entre cada pico detectado. 
#print(indexes) 
print(x[indexes], y[indexes]) #Imprime las coordenadas del pico
pyplot.figure(figsize=(10,6)) #Declara una figura con el tamaño establecido
pplot(x, y, indexes) #Plotea la figura usando los datos; aquí se plotea el pico con pplot, que es el plot de peakutils
pyplot.title('Primer estimado') #Le pone titulo a la figura
pyplot.grid() #Activa la cuadricula en la figura
```

```python
#Sin el arreglo de numeros random, no hay ruido
centers2 = (30.0, 70.0)
x2 = np.linspace(0, 500, 150)
y2 = (peakutils.gaussian(x2, 5, centers2[0], 3) +
     peakutils.gaussian(x2, 7, centers2[1], 10))
pyplot.figure(figsize=(10,6))
pyplot.plot(x2,y2)
pyplot.title('Datos sin ruido, solo las funciones Gaussianas')
pyplot.grid()
```

```python
#Encontrando el pico sin ruido. La aproximacion no estan buena sin ruido
indexes2 = peakutils.indexes(y, thres=0.3, min_dist=11) #A partir de 11 solo detecta el pico más grande (para este caso)
print(x[indexes2], y[indexes2])
pyplot.figure(figsize=(10,6))
pplot(x2, y2, indexes2)
pyplot.title('Primer estimado, sin ruido')
pyplot.grid()
```
